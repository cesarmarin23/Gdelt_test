```python
import pandas as pd 
import numpy as np
import glob
```

```python
path = r'/Users/cesarmarin/Downloads/gdelt_test' # use your path
all_files = glob.glob(path + "/*.csv")
```

```python
all_files
```

```python
li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=None, sep='\t',encoding='ISO-8859-1')
    li.append(df)

df= pd.concat(li, axis=0, ignore_index=True)
```

```python
column_names = ['GKGRECORDID','V2_DATE','V2SOURCEIDENTIFIER','V2SOURCECOMMONNAME','V2DOCUMENTURL','V1COUNTS','V2_COUNTS','V1THEMES',
'V2ENHANCEDTHEMES','V1LOCATIONS','V2ENHANCEDLOCATIONS','V1PERSONS','V2ENHANCEDPERSONS','V1ORGANIZATIONS',
'V2ORGANIZATIONS','V1_TONE','V2_1ENHANCEDDATES','V2GCAM','V2_1SHARINGIMAGE','V2_1RELATEDIMAGES','V2_1SOCIALIMAGEEMBEDS',
'V2_1SOCIALVIDEOEMBEDS','V2_1QUOTATIONS','V2_1ALLNAMES','V2_AMOUNTS','V2_TRANSLATIONINFO','V2EXTRASXML']
```

```python
df.columns = column_names
df
```

```python
df['Company'] = pd.np.where(df['V2ORGANIZATIONS'].str.contains("Exxon").fillna(False),'Company1',
                            (np.where(df['V2ORGANIZATIONS'].str.contains("Toyota").fillna(False),'Company2',' ')))
 
```

```python
df_companies = df[(df['Company'] == 'Company2')| (df['Company']=='Company1')]
```

```python
df_companies.to_excel('Companies_1.xlsx')
```
